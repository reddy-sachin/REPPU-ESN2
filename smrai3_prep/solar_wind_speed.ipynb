```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime 
```

```python
df = pd.read_csv('/home/sachin/Documents/NIPR/Research/Data/OMNI/1-min/smrai3/omni_hro2_1min_2020-2024.csv')
df['dt'] = pd.to_datetime(df['dt'])
daily_omni = df.resample('27D', on='dt').mean().reset_index()
daily_omni
```

```python
pastel_colors = sns.color_palette("pastel")
pastel_array = np.array(pastel_colors)
pastel_colors
```

```python
fig, ax = plt.subplots(2,1, figsize=(7,3.7), sharex=True)

sns.lineplot(data=daily_omni, x='dt', y='flow_speed', ax=ax[0], color='black',drawstyle='steps-pre', linewidth=1.2)
sns.lineplot(data=daily_omni, x='dt', y='BZ_GSE', ax=ax[1],color='black',drawstyle='steps-pre', linewidth=1.2)

alpha = 0.25
ax[0].axvspan(datetime.datetime(2020,8,1), datetime.datetime(2021,12,1), color=pastel_array[0], alpha=alpha) #train 1
ax[0].axvspan(datetime.datetime(2021,12,2), datetime.datetime(2022,4,20), color=pastel_array[1], alpha=alpha) #val 
ax[0].axvspan(datetime.datetime(2022,4,21), datetime.datetime(2023,7,12), color=pastel_array[0], alpha=alpha) #train 2
ax[0].axvspan(datetime.datetime(2023,7,12), datetime.datetime(2023,12,1), color=pastel_array[2], alpha=alpha) #test
ax[0].axvspan(datetime.datetime(2023,12,1), datetime.datetime(2024,7,30), color=pastel_array[0], alpha=alpha) #train 3

ax[0].grid(axis='y', linestyle='-', linewidth=0.5)
ax[0].set_xlabel(' ')
ax_y_ticks = np.arange(350, 550, 50)
ax[0].set_yticks(ax_y_ticks)
#ax[0].set_ylim([45,210])


ax[1].axvspan(datetime.datetime(2020,8,1), datetime.datetime(2021,12,1), color=pastel_array[0], alpha=alpha) #train 1
ax[1].axvspan(datetime.datetime(2021,12,2), datetime.datetime(2022,4,20), color=pastel_array[1], alpha=alpha) #val 
ax[1].axvspan(datetime.datetime(2022,4,21), datetime.datetime(2023,7,12), color=pastel_array[0], alpha=alpha) #train 2
ax[1].axvspan(datetime.datetime(2023,7,12), datetime.datetime(2023,12,1), color=pastel_array[2], alpha=alpha) #test
ax[1].axvspan(datetime.datetime(2023,12,1), datetime.datetime(2024,7,30), color=pastel_array[0], alpha=alpha) #train 3



#add annotations of train, val, test
height = -10
#ax[0].text(datetime.datetime(2013,3,1), height, 'Train', fontsize=10, color='black')
#ax[0].text(datetime.datetime(2017,5,1), height, 'Test', fontsize=10, color='black')
#ax[0].text(datetime.datetime(2019,6,15), height, 'Train', fontsize=10, color='black')
#ax[0].text(datetime.datetime(2020,11,15), height, 'Val', fontsize=10, color='black')
#ax[0].text(datetime.datetime(2022,11,1), height, 'Train', fontsize=10, color='black')

ax_y_ticks = np.arange(-1, 1.1, .5)
ax[1].set_yticks(ax_y_ticks)
#ax[1].set_ylim([0,3.2])

ax[1].set_xlabel(' ')
#ax[1].set_ylabel('Kp', labelpad=15)

#loop through the axes and remove the spines
for i in ax:
    i.spines['top'].set_visible(False)
    i.spines['right'].set_visible(False)
    i.spines['left'].set_visible(False)
    i.spines['bottom'].set_visible(False)
    i.grid(axis='y', linestyle='-', alpha=0.7)


plt.tight_layout()
plt.savefig('F107_Kp_2009-2024.png', dpi=500)
```
