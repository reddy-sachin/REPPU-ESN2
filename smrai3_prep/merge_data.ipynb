```python
def add_imf_sw(ds, res, feat, lon,hemi):
    imf_df = pd.read_csv('/home/sachin/Documents/NIPR/Research/Data/OMNI/1-min/smrai3/rtsw_2min_2020-2024.csv') #rtsw
    #imf_df = pd.read_csv('/home/sachin/Documents/NIPR/Research/Data/OMNI/1-min/smrai3/omni_hro2_1min_2020-2024.csv') #omni_hro2

    imf_df['dt'] = pd.to_datetime(imf_df['dt']) #convert to datetime
    imf_df = imf_df.set_index('dt') #set datetime as index

    imf_df = imf_df.resample(f'{res}min').mean()
    imf_df = imf_df.interpolate(method='linear')
    
    imf_ds = xr.Dataset.from_dataframe(imf_df) #convert to xarray dataset

    merged = xr.merge([ds, imf_ds], join='inner') #merge the datasets

    merged.to_netcdf(f'/run/media/sachin/0fa21ddb-f70c-4238-9cf4-705e0360f1c1/NICT_Data/{hemi}/multi/{hemi}_2020-2024_{feat}_feats_{lon}_40_{res}min_rtsw.nc') #save the merged dataset

    return merged

ml_ds = add_imf_sw(merged_ds, 2, feat, 140, hemi)
ml_ds
```

```python
def nan_count(ds, var):
    #count number of NaNs in a data variable
    con_data = ds[var].values
    nan_count = np.count_nonzero(np.isnan(con_data))
    nan_ratio = nan_count / con_data.size
    not_nan = con_data.size - nan_count
    return not_nan, nan_count, np.round(nan_ratio, 2)

nan_count(ml_ds, feat)
```
